```python
import pandas as pd
import numpy as np
```

```python
dict1 = {
    'From_To': [
        'LoNDon_paris',
        'MAdrid_miLAN',
        'londON_StockhOlm',
        'Budapest_PaRis',
        'Brussels_londOn'
    ],
    'FlightNumber': [
        10045,
        np.nan,
        10065,
        np.nan,
        10085
    ],
    'RecentDelays': [
        [23, 47],
        [],
        [24, 43, 87],
        [13],
        [67, 32]
    ],
    'Airline': [
        'KLM(!)',
        '<Air France> (12)',
        '(British Airways. )',
        '12. Air France',
        '"Swiss Air"'
    ]
}
```

```python
df = pd.DataFrame(dict1)
df
```

# Problem-1

```python
df['FlightNumber']=df['FlightNumber'].fillna(value=pd.Series([10045,10055,10065,10075,10085]))
```

```python
df
```

```python
df.dtypes
```

```python
df = df.astype({'FlightNumber':int})
```

```python
df
```

# Problem-2

```python
df2 = df.From_To.str.split(pat='_',expand=True)
df2=df2.rename(columns={0:'From',1:'To'})
df2
```

# Problem-3

```python
df2.From = df2.From.apply(lambda x: x.capitalize())
df2.To = df2.To.apply(lambda x: x.capitalize())
df2
```

# Problem-4

```python
df3 = pd.concat([df2,df.drop(columns=['From_To'])],axis=1)
df3
```

# Problem-5

```python
df_temp = pd.DataFrame(df3.RecentDelays.to_list())
df_temp
```

```python
df_temp=df_temp.rename(columns={0:'delay_1',1:'delay_2',2:'delay_3'})
df_temp
```

```python
df4 = pd.concat(
    [
        df3.drop(columns=['RecentDelays','Airline']),
        df_temp,
        df3['Airline']
    ],
    axis=1)
df4
```

```python
df4.Airline = df4.Airline.str.replace('[^a-zA-Z]', '')
df4
```
